```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

interval = np.arange(0.00, 1.00, 0.05)

size = len(interval)
```

```python
f = open('/home/dreadnought/Documents/IC2017/data_medias.txt', 'r')
```

```python
lines = [i.split() for i in f.read().split('\n')]
f.close()
```

```python
lines = lines[:-1]
```

```python
TN = [float(i[0]) for i in lines]
FN = [float(i[1]) for i in lines]
```

```python
names = ['ExG', 'ExGR', 'CIVE', 'VEG', 'WI', 'NGRDI']
```

```python
plt.figure(figsize=(12,8))
for i in range(6):
    start = i * size
    end = start + size
    print(TN[start:end])
    print(FN[start:end])
    if FN[start + 6] < TN[start + 6]:
        plt.plot(FN[start:end],TN[start:end], label = names[i])
    else:
        plt.plot(TN[start:end],FN[start:end], label = names[i])
plt.legend()
plt.xlabel('FPR')
plt.ylabel('TPR')
#plt.yscale('log', basey = 2)
```

```python
TP = [float(i[2]) for i in lines]
FP = [float(i[4]) for i in lines]
```

```python
plt.figure(figsize=(15,10))
for i in range(6):
    plt.plot(FP[i::6],TP[i::6], label = lines[i][0])
plt.legend()
```

```python
plt.figure(figsize=(15,10))

for i in range(4):
    plt.subplot(2,2,i + 1)
    plt.scatter(FN[i::6], TN[i::6])
    plt.title(lines[i][0])
    plt.xlabel('FP')
    plt.ylabel('TP')
    
plt.tight_layout()
```
