```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.util.shape import view_as_blocks

from sklearn.metrics import f1_score, roc_curve, roc_auc_score, precision_recall_curve, accuracy_score

from scipy.misc import imresize
```

```python
data = pd.read_csv('../IC2017_DATA/PIXELdataset.csv')
```

```python
data.columns
```

```python
seg = data.query('IMG == 26')
```

```python
exg = seg['ExG']
```

```python
gt = seg['GT']
```

```python
x = exg.values.reshape(512, 512)
```

```python
plt.imshow(x)
```

```python
r = 32
s = 512 // r

y_block = view_as_blocks(x, (s, s))
y_stat = y_block.mean(axis = (2, 3))
y_rsz = imresize(y_stat, (512, 512))
plt.imshow(y_rsz)
```

```python
# %%time
resizing_factors = [8, 16, 32, 64]
functions = [np.mean, np.max, np. min, np.std]
nr = len(resizing_factors)
nf = len(functions)

out = np.zeros((512 * 512, 1 + nr * nf))
for i, r in enumerate(resizing_factors):
    s = 512 // r
    x_block = view_as_blocks(x, (s, s))
    for j, f in enumerate(functions):
        print(i * nr + j, r, f)
        x_stat = f(x_block, axis = (2, 3))
        x_rsz = imresize(x_stat, (512, 512))
        out[: , i * nf + j] = x_rsz.ravel()
        
out[: , nr * nf] = x.ravel()
```

```python
out[1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
est = LogisticRegression(penalty='l1')
```

```python
est.fit(out, gt)
```

```python
pred = est.predict_proba(out)[:, -1].reshape(512, 512)
```

```python
plt.imshow(pred)
plt.show()
plt.imshow(gt.values.reshape(512, 512))
plt.show()
plt.imshow(x)
```
